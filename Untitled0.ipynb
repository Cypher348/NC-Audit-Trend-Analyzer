```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load file
df = pd.read_excel('/content/drive/MyDrive/Audit_nc_project/nc_audit_data.xlsx', sheet_name='NC_Data')

# Sanity check
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nNC Category Distribution:\n", df['NC_Category'].value_counts())
print("\nSeverity Distribution:\n", df['Severity'].value_counts())
print("\nStatus Distribution:\n", df['Status'].value_counts())
```

```python
# Convert date column
df['Audit_Date'] = pd.to_datetime(df['Audit_Date'])

# Fill missing closure days with 0 for open NCs
df['Closure_Days'] = df['Closure_Days'].fillna(0)

# Audit type short label
df['Audit_Level'] = df['Audit_Type'].str[:2]

print("Date range:", df['Audit_Date'].min(), "to", df['Audit_Date'].max())
print("Preprocessing done. Shape:", df.shape)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Step 1 - Create canvas
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Step 2 - Count NCs per category
cat_counts = df['NC_Category'].value_counts()

# Step 3 - Draw colored bars
colors = ['#E74C3C','#E67E22','#F1C40F','#2ECC71','#3498DB','#9B59B6']
axes[0,0].bar(cat_counts.index, cat_counts.values, color=colors)

# Step 4 - Add title and labels
axes[0,0].set_title('NC Count by Category', fontweight='bold')
axes[0,0].set_xlabel('Category')
axes[0,0].set_ylabel('Count')
axes[0,0].tick_params(axis='x', rotation=30)

# Step 5 - Add count numbers on top of bars
for i, v in enumerate(cat_counts.values):
    axes[0,0].text(i, v+1, str(v), ha='center', fontweight='bold')

# Step 6 - Render
plt.tight_layout()
sev_counts=df['Severity'].value_counts()
axes[0,1].pie(sev_counts.values, labels=sev_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
axes[0,1].set_title('NC Severity Distribution', fontweight='bold')
#station wise nc count
station_counts=df['Station_Name'].value_counts().head(10)
axes[1,0].barh(station_counts.index, station_counts.values,color='#3498DB')
axes[1,0].set_title('NC Count by Station', fontweight='bold')
axes[1,0].set_xlabel('Count')
axes[1,0].set_ylabel('Station')
# NC status count breakdown
status_counts=df['Status'].value_counts()
axes[1,1].bar(status_counts.index,status_counts.values,color=['#2ECC71','#E74C3C','#E67E22'])
axes[1,1].set_title('NC Status breakdown', fontweight='bold')
axes[1,1].set_xlabel('Count')
for i, v in enumerate(status_counts.values):
  axes[1,1].text(v+1, i, str(v), ha='center', fontweight='bold')
  plt.savefig('/content/drive/MyDrive/Audit_nc_project/eda_dashboard.png',
            dpi=150, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
Monthly=df.groupby('Month')['NC_ID'].count()
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']
Monthly = Monthly.reindex(month_order)
print(Monthly)
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(range(len(Monthly)), Monthly.values,
        marker='o', color='#E74C3C', linewidth=2, markersize=8)
ax.set_xticks(range(len(Monthly)))
ax.set_xticklabels([m[:3] for m in month_order], rotation=30)
ax.set_title('Monthly NC Trend Line', fontweight='bold', fontsize=14)
ax.set_ylabel('NC Count')
ax.grid(axis='y', alpha=0.4)
plt.tight_layout()
plt.show()
```
